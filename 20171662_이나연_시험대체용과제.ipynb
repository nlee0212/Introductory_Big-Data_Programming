```python
%%writefile OOP_v2.py

import time

def checkTime(func):
    def new_func(*args):
        print("\n------------------------------------------------\n")
        startsec = time.time()
        func(*args)
        print("\n실행시간은:",time.time()-startsec)
        print("\n------------------------------------------------\n")
    return new_func

def is_leap(year):
    if year%4 == 0:
        if year%400 == 0 or year%100 != 0:
            return 1
        else:
            return 0

class Person(object):
    datenum = 0
    def __init__(self):
        self._name = ""
        self._year = 0
        self._month = 0
        self._day = 0
        self.day_num = 0
        self.day_name = {
            1: "월요일", 2: "화요일", 3: "수요일", 
            4: "목요일", 5: "금요일", 6: "토요일", 
            0: "일요일"
        }
        self.isleap = 0
        self.liveday = 0
        Person.datenum+=1
        
    @property
    def year(self):
        return self._year
    
    @year.setter
    def year(self, year):
        self._year = year
        
    @property
    def month(self):
        return self._month
    
    @month.setter
    def month(self, month):
        self._month = month
        
    @property
    def day(self):
        return self._day
    
    @day.setter
    def day(self, day):
        self._day = day
        
    @property
    def name(self):
        return self._name
    
    @name.setter
    def name(self, name):
        self._name = name
                
    def cal_day_num(self):
        self.isleap = is_leap(self.year)
        for i in range(1,self.year):
            if(is_leap(i)):
                self.day_num+=366
            else:
                self.day_num+=365
            
        for m in range(1,self.month):
            if m<8:
                if m%2==1:
                    self.day_num+=31
                elif m!= 2:
                    self.day_num+=30
                elif (self.isleap):
                    self.day_num+=29
                else: 
                    self.day_num+=28
            else:
                if m%2==0:
                    self.day_num+=31
                else:
                    self.day_num+=30
                
        self.day_num+=self.day
        #print(self.day_num)
        
    def calc_day_name(self):
        print((self.day_name).get((self.day_num)%7)+"에 태어났습니다.")
        
    def calc_live_day(self):
        now_day = self.now_day_num()
        self.cal_day_num()
        print("생후 {}일차".format(now_day - self.day_num + 1))
        self.liveday = now_day - self.day_num + 1
        
    @staticmethod    
    def now_day_num():
        t = time.localtime()
        now = Person()
        now.year = t.tm_year
        now.month = t.tm_mon
        now.day = t.tm_mday
        now.cal_day_num()
        Person.datenum-=1
        return now.day_num
    
    def find_zodiac(self):
        if(self.month==1 and self.day>=20) or (self.month==2 and self.day<=18):
            return "물병자리"
        elif(self.month==2 and self.day>=19) or (self.month==3 and self.day<=20):
            return "물고기자리"
        elif(self.month==3 and self.day>=21) or (self.month==4 and self.day<=19):
            return "양자리"
        elif(self.month==4 and self.day>=20) or (self.month==5 and self.day<=20):
            return "황소자리"
        elif(self.month==5 and self.day>=21) or (self.month==6 and self.day<=21):
            return "쌍둥이자리"
        elif(self.month==6 and self.day>=22) or (self.month==7 and self.day<=22):
            return "게자리"
        elif(self.month==7 and self.day>=23) or (self.month==8 and self.day<=22):
            return "사자자리"
        elif(self.month==8 and self.day>=23) or (self.month==9 and self.day<=23):
            return "처녀자리"
        elif(self.month==9 and self.day>=24) or (self.month==10 and self.day<=22):
            return "천칭자리"
        elif(self.month==10 and self.day>=23) or (self.month==11 and self.day<=22):
            return "전갈자리"
        elif(self.month==11 and self.day>=23) or (self.month==12 and self.day<=24):
            return "사수자리"
        else:
            return "염소자리"
        
class Employee(Person):
    emp_num = 0
    def __init__(self, team, salary):
        Employee.emp_num+=1
        self.ID = ""
        self.salary = salary
        self.team = team
        self.teamname = ""
        self.setID()
        self.setteamname()
        
    def setID(self):
        self.ID += str(time.localtime().tm_year)
        self.ID += str(self.team)
        self.ID += str(Employee.emp_num)
        
    def setteamname(self):
        if self.team==1:
            self.teamname = "인사팀"
        elif self.team == 2:
            self.teamname = "마케팅팀"
        elif self.team == 3:
            self.teamname = "재무팀"
        else:
            self.teamname = "기획팀"
    
    @classmethod
    def get_emp_num(cls):
        print("\n{}명의 면접자 중 {}명을 고용하였습니다.\n".format(Person.datenum, cls.emp_num))
        
    
        
def saveInfo():
    people_num = int(input("\n몇 명의 정보를 저장하겠습니까? "))
    for i in range(people_num):
        print("\n{}번째 입력".format(i+1))
        getInfo()
        person = save_person[Person.datenum-1]
        person.calc_live_day()
        person.calc_day_name()
        if(person.isleap): print("태어난 년도는 윤년입니다.")
        
def getInfo():
    name = input("이름을 입력하시오: ")
    bday = input("생일을 입력하시오 (ex. 1998, 1, 4): ")
    
    person = Person()
    person.name = name
    arr = bday.split(', ')
    person.year = int(arr[0])
    person.month = int(arr[1])
    person.day = int(arr[2])
    save_person.append(person)
    hire_employee.append(0)
    
@checkTime
def showInfo(index):
    for i in range(index):
        cur = save_person[i]
        print("이름: "+cur.name)
        print("생일: "+str(cur.year)+"년 "+str(cur.month)+"월 "+str(cur.day)+"일")
        print("별자리: "+cur.find_zodiac()+"\n")

def responsibility(person):
    team = int(input("\n어떤 직무팀에 배정하시겠습니까?\n1: 인사팀, 2: 마케팅팀, 3: 재무팀, 4: 기획팀 (숫자를 입력하시오.)\n"))
    if team<1 or team>4:
        print("1~4의 숫자를 입력하시오.")
        return
    salary = input("해당 직원의 연봉은 얼마입니까?\n")
    if int(salary) < 0 or salary.isdigit()==False:
        print("연봉은 0 이상의 금액으로 입력하시오.")
        return
    else:
        emp = Employee(team, int(salary))
        emp.name = person.name
        emp.year = person.year
        emp.month = person.month
        emp.day = person.day
        emp.liveday = person.liveday
        save_employee.append(emp)

@checkTime
def showEmpInfo(index):
    Employee.get_emp_num()
    young_day = 1000000
    for i in range(index):
        emp = save_employee[i]
        if(emp.liveday < young_day):
            young_name = emp.name
            young_day = emp.liveday
        print("\n이름: "+emp.name)
        print("ID: "+emp.ID)
        print("직무: "+emp.teamname)
        print("연봉: "+str(format(emp.salary, ','))+"원")
    if index>0:
        print("\n나이가 제일 어린 직원은 {}입니다.".format(young_name))
        
if __name__ == "__main__":
    save_person = []
    save_employee = []
    hire_employee=[]
    print("***직원 관리 메뉴에 오신 걸 환영합니다.***\n")
    while True:
        menu = input("\n1. 면접자\n2. 면접자 목록\n3. 입사자 직무배정\n4. 직원 목록\n5. 프로그램 종료\n")
        if menu.isdigit()==False:
            print("1~5사이의 숫자를 입력하시오.")
            continue
        menu = int(menu)
        if menu == 1:
            saveInfo()
        elif menu == 2:
            if Person.datenum == 0:
                print("면접자를 등록한 후 해당 메뉴를 사용 바랍니다.")
                continue
            showInfo(Person.datenum)
        elif menu == 3:
            for i in range(Person.datenum):
                if hire_employee[i]==1:
                    print("\n{}의 직무 배정을 이미 완료했습니다.".format(save_person[i].name))
                    continue
                hire = input("\n{}의 직무를 배정하시겠습니까? (Y/N) ".format(save_person[i].name))
                if hire == "Y":
                    responsibility(save_person[i])
                    hire_employee[i]=1
                elif hire == "N":
                    hire_employee[i]=0
                    continue
                else:
                    print("Y 또는 N을 입력하시오.")
        elif menu == 4:
            showEmpInfo(Employee.emp_num)
        elif menu == 5:
            break
        else:
            print("1 ~ 5의 숫자를 입력하시오.")
    print("\n***직원 관리 프로그램을 종료합니다.***")
```

```python
%matplotlib inline
from pylab import *

x = linspace(-pi,pi,256)
c = cos(x)

plot(x,c)
xticks([-pi, -pi/2, 0, pi/2, pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
yticks([-1,0,1])
```

```python
%matplotlib inline
from pylab import *

x = linspace(-pi,pi,256)
c = cos(x)
s = sin(x)

plot(x,c)
plot(x,s)
xticks([-pi, -pi/2, 0, pi/2, pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
yticks([-1,0,1])
```
